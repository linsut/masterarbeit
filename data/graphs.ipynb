```python
import matplotlib.pyplot as plt
```

```python
# Sentiment Analysis
TextBlobAcc=[0.530,0.257,0.661]
ELMoLogAcc=[0.838,0.810,0.]
ELMoLstmAcc=[0.820,0.815,0.]
ULMAcc=[0.824,0.885,0.975]
RobertaAcc=[0.786,0.,0.]
SimpleRobertaAcc=[0.887,0.,0.]
XLNetAcc=[0.779,0.662,0.]
SimpleXLNetAcc=[0.,0.,0.]
SimpleGPTAcc=[0.,0.,0.]
sentiment_sets=['Apple Sentiment', 'US Airlines', 'T4SA']


plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(15,10))
plt.plot(sentiment_sets,TextBlobAcc, label='TextBlob', marker='o')
plt.plot(sentiment_sets,ELMoLogAcc, label='ELMo Logistic', marker='o')
plt.plot(sentiment_sets,ELMoLstmAcc, label='ELMo LSTM', marker='o')
plt.plot(sentiment_sets,ULMAcc, label='ULMFit', marker='o')
plt.plot(sentiment_sets,RobertaAcc, label='RoBERTa', marker='o')
plt.plot(sentiment_sets,SimpleRobertaAcc, label='RoBERTa simple', marker='o')
plt.plot(sentiment_sets,XLNetAcc, label='XLNet', marker='o')
plt.plot(sentiment_sets,SimpleXLNetAcc, label='XLNet simple', marker='o')
plt.plot(sentiment_sets,SimpleGPTAcc, label='GPT simple', marker='o')
plt.ylabel('Genauigkeit')
plt.legend()
plt.show()
```
